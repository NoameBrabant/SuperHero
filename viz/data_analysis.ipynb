```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import json
```

```python
df = pd.read_csv('./sources/lecsv/characters_df3.csv')
df.head(5)
```

```python
df["Mean_Stats"] = sum([df["CON"], df["CHA"], df["WIS"], df["INT"], df["STR"], df["DEX"]]) / 6
```

```python
df["Mean_Stats"] = df["Mean_Stats"].apply(lambda x: round(x, 2))
```

```python
def get_medium_image(image : str) -> str:
	data = json.loads(image.replace("'", "\""))
	return data["super_url"]

df["image_medium"] = df["Image"].apply(get_medium_image)
df["image_medium"].head(5)
```
